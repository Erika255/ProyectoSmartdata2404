```python
%python
dbutils.widgets.removeAll()
```

```python
%python
dbutils.widgets.text("storageName", "adlsproyecto2404")
dbutils.widgets.text("catalogo", "catalog_clinica")
```

```python
%python
storageName = dbutils.widgets.get("storageName")
catalogo = dbutils.widgets.get("catalogo")
```

### Eliminacion tablas Bronze

```python
%sql
DROP TABLE IF EXISTS ${catalogo}.bronze.citas;
DROP TABLE IF EXISTS ${catalogo}.bronze.pacientes;
DROP TABLE IF EXISTS ${catalogo}.bronze.medicos;
DROP TABLE IF EXISTS ${catalogo}.bronze.especialidades;
DROP TABLE IF EXISTS ${catalogo}.bronze.atenciones;
DROP TABLE IF EXISTS ${catalogo}.bronze.pagos;
```

```python
%python
dbutils.fs.rm(f"abfss://bronze@{storageName}.dfs.core.windows.net/citas", True)
dbutils.fs.rm(f"abfss://bronze@{storageName}.dfs.core.windows.net/pacientes", True)
dbutils.fs.rm(f"abfss://bronze@{storageName}.dfs.core.windows.net/medicos", True)
dbutils.fs.rm(f"abfss://bronze@{storageName}.dfs.core.windows.net/especialidades", True)
dbutils.fs.rm(f"abfss://bronze@{storageName}.dfs.core.windows.net/atenciones", True)
dbutils.fs.rm(f"abfss://bronze@{storageName}.dfs.core.windows.net/pagos", True)
```

### Eliminacion tablas Silver

```python
%sql
DROP TABLE IF EXISTS ${catalogo}.silver.citas_detalle;
DROP TABLE IF EXISTS ${catalogo}.silver.atenciones_detalle;
```

```python
%python
dbutils.fs.rm(f"abfss://silver@{storageName}.dfs.core.windows.net/citas_detalle", True)
dbutils.fs.rm(f"abfss://silver@{storageName}.dfs.core.windows.net/atenciones_detalle", True)
```

### Eliminacion tablas Golden

```python
%sql
DROP TABLE IF EXISTS ${catalogo}.golden.desercion_por_especialidad;
DROP TABLE IF EXISTS ${catalogo}.golden.desercion_por_seguro;
DROP TABLE IF EXISTS ${catalogo}.golden.ingresos_por_especialidad;
DROP TABLE IF EXISTS ${catalogo}.golden.pacientes_recurrentes;
```
